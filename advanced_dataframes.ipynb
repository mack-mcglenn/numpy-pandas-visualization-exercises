```python
# Exercise 2.8: Check out your column names and perform any cleanup you may want on them.

mdf = mdf.rename(columns={'manufacturer': 'Manufacturer', 'model':'Model',
                 'displ':'l', 'year':'Year', 'cyl':'Cyl',
                 'trans':'Trans', 'drv':'Drv', 
                 'cty':'City', 'hwy':'Highway', 
                 'fl':'Fuel', 'class':'Class'})



mdf
```

```python
# Exercise 2.10: How many different manufacturers are there?

mdf['Manufacturer'].unique().size
```

```python
#  Exercise 2.12: Create a column named mileage_difference like you did in the DataFrames
# exercises; this column should contain the difference 
# between highway and city mileage for each car.

mdf['Mileage Difference'] = mdf['Highway'] - mdf['City']
mdf
```

```python
# Exercise 2.13: Create a column named average_mileage like you did in the
# DataFrames exercises; this is the mean of the city and highway mileage.

mdf['Avg Mileage'] = mdf[['City', 'Highway']].mean(axis=1)
```

```python
mdf
```

```python
#Ignore this, had to drop a duplicate column

mdf = mdf.drop(columns=['automatic'])
mdf
```

```python
# Exercise 2.14: Create a new column on the mpg dataset named is_automatic that holds boolean
# values denoting whether the car has an automatic transmission.

mdf['Automatic'] = mdf['Trans'].str.contains('auto')
mdf
```

```python
# Exercise 2.15: Using the mpg dataset, find out which which manufacturer 
# has the best miles per gallon on average?

mdf.groupby('Manufacturer').mean()['Avg Mileage'].nlargest(1)
```

```python
# Exercise 2.16: Do automatic or manual cars have better miles per gallon?
# A: Manual

mdf.groupby('Automatic').mean()['Avg Mileage']
```
