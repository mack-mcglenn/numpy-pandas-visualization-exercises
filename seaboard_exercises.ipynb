```python
# Exercise 3.1 Using the chipotle dataset from the previous exercise, create a bar chart
# that shows the 4 most popular items and the revenue produced by each.


def get_db_url(username, password, host, db):
  
    url = f'mysql+pymysql://{username}:{password}@{host}/{db}'
    return url
```

```python
gross = get_db_url(username, password, host, 'chipotle')
```

```python
gross_guac= pd.read_sql('select * from orders', gross)
gross_guac.head(21)
```

```python
# I need to convert my dtype for 'item_price' into a float
gross_guac['item_price']=gross_guac['item_price'].str.replace('$', '')
gross_guac['item_price'] = gross_guac['item_price'].astype(float)
gross_guac['item_price']
```

```python
# Then, I group by 'item_name' and use my sum function to get a total float value for my
# item price/int value for my quantity
top_sellers= gross_guac.groupby('item_name').sum()
top_sellers
```

```python
# Next, I neet to organize my table to show my highest quantity items at the top
# A: Top 4 sellers: Chicken Bowl, Chicken Burrito, Chips & Guac, Steak Burrito

dry_chicken = top_sellers.sort_values(by='quantity', ascending=False).head(4)
dry_chicken
```

```python
# To find my items bringing in the most revenue, I use the same function as the last block,
# but I sort by 'item_price' rather than 'quantity'

# A: Top sellers by revenue: Chicken Bowl, Chicken Burrito, Steak Burrito, Steak Bowl

cold_stale_chips= top_sellers.sort_values(by='item_price', ascending=False).reset_index().head(4)
cold_stale_chips
```
