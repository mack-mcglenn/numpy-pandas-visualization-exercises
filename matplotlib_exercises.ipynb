```python
# Exercise 1: Use matplotlib to plot the following equation:
# y=x2−x+2

# You'll need to write the code that generates the x and y points.

# Add an anotation for the point 0, 0, the origin.

x = range(0, 100)
y= [x**2 - x + 2 for x in x] 

plt.plot(x, y)
plt.annotate('origin', xy=(0,0))
plt.title("'y = x^2 - x + 2'")
plt.show()
```
